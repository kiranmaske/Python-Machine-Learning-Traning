```python
import numpy as np
ar1=np.arange(1,18)
print(ar1)
```

```python
ar2=np.arange(1,18,2)
#  number 2 indicates it print 
print(ar2)
```

```python
# we want change datatype of array 
ar3=np.arange(1,18,2,dtype=float)
print(ar3)
```

```python
x=np.linspace(10,20)
#  num is not include & num default is 50 so the range 
# 10 to 20 evaly spaced sample with 50 different object
print(x)
```

```python
x=np.linspace(10,20,num=5)
#  num value taken 5 so the range is evaluy 10 to 20 is 5
print(x)
```

```python
#end point set to false 
x=np.linspace(10,20,5,endpoint =False)

print(x)
```
