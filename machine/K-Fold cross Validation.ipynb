```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
df=load_digits()
```

```python
df.data
```

```python
df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.data,df.target,test_size=0.3)
```

```python
#Apply algorithem 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
lr=LogisticRegression(solver='liblinear',multi_class='ovr')
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
#SVC
# gamma is a parameter for non linear perspective 

svm = SVC(gamma='auto')
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
# Random forest 
rf = RandomForestClassifier(n_estimators=40)
rf.fit(x_train,y_train)
rf.score(x_test,y_test)
```

```python
#k-Fold Cross validation on the logistic regression
from sklearn.model_selection import cross_val_score
```

```python
score_lr = cross_val_score(
    LogisticRegression(solver='liblinear', multi_class='ovr'),
    df.data,
    df.target,
    cv=5);
print(score_lr)
print("Avg:",np.average(score_lr))
```

```python
score_svm = cross_val_score (SVC(gamma='auto'),df.data,df.target,cv=3)

   
print(score_svm)
print("Avg:",np.average(score_svm))
```

```python
score_rf = cross_val_score( RandomForestClassifier(n_estimators=40),df.data,df.target,cv=5);
print(score_rf)
print("Avg:",np.average(score_rf))
```
