```python
import pandas as pd 
```

```python
employees=[('pravin',28,'Delhi'),
           ('satish',32,'Varanasi'),
           ('Aditya',25,'Mumbai'),
           ('Saumya',32,'Delhi'),
           ('Saumya',32,'Delhi'),
           ('Saumya',32,'Delhi'),
           ('Saumya',32,'Mumbai'),
           ('Aditya',25,'Dehradun'),
           ('sema',32,'Delhi')]

df = pd.DataFrame(employees,columns=['Name','Age','City'])
df
```

```python
duplicate=df[df.duplicated()]
print("Duplicate Rows:")
duplicate
```

```python
duplicate=df[df.duplicated('Name')]
print("Duplicate Rows based on 'Name'")
duplicate
```

```python
duplicate=df[df.duplicated('City')]
print("Duplicate Rows based on 'City'")
duplicate
```

```python
duplicate=df[df.duplicated(['Name','Age'])]
print("Duplicate Rows based on Name and Age:")
duplicate
```

```python
duplicate_sort=df[df.duplicated(['Name','Age'],keep=False)].sort_values('City')
print("Duplicate Rows based on Name and Age(sorted):")
duplicate_sort
```
