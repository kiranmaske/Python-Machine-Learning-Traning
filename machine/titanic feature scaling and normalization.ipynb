```python
import numpy as np
import pandas as pd
import seaborn as sns 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
pd.read_csv('titanic.csv')
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df2=df[['survived',"pclass",'age','parch']]
df2.head()
```

```python
df2.isna().sum()
```

```python
# now first clean the data 
df3=df2.fillna(df2.mean())
```

```python
df3.isnull().sum()
```

```python
#  convert data in matrix and vector form 
#  Bez independent variable in matrix from 
# dependent variable in vector from 
```

```python
x= df3.drop("survived",axis=1)
#To show the matrix we defined X(capital)
# we drop "survived " because it is depend variable & in matrix we defined 
y=df3["survived"]
# y is use to show vector (for dependent variable)
print('Shape of X',x.shape)
print('Shape of Y',y.shape)
```

```python
#from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=51)
print('Shape of X_train',x_train.shape)
print('Shape of Y_train',y_train.shape)
print('Shape of X_test',x_test.shape)
print('Shape of Y_test',y_test.shape)
```

```python
sc=StandardScaler()
sc.fit(x_train)
```

```python
#Attribute 
sc.mean_
```

```python
sc.scale_
```

```python
x_train.describe()  #diff feature numrical distribution likke mean media 
```

```python
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
```
