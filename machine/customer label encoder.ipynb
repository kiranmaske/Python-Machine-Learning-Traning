```python
df=pd.read_csv('Customer.csv')
```

```python
df.info()
```

```python
df1=df.iloc[:,2:4]
df1.head()
```

```python
df1['Review'].value_counts()
```

```python
df1['Education'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['School','UG','PG'],['Poor','Average','Good']])

df1_oe= oe.fit_transform(df1)
df1_oe
```

```python
import numpy as np
type(df1_oe)
np.ndarray
df1=pd.DataFrame(df1_oe,columns=['Education','Review'])
df1.head()
```

```python
df.update(df1)
#  now we update our ordinal categori variable to number format using update method 
# but it you see 'review' & Education coloumn it shown f;oat values 
df.head()
```

```python
df['Review']=df["Review"].astype(int)
df['Education']=df["Education"].astype(int)
```

```python
df.info()
```

```python
#lable encoder 
df2=df[['Gender','Purchased']]
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df2["Gender_encoder"]=le.fit_transform(df2["Gender"])
df2["Purchased_encoder"]=le.fit_transform(df2["Purchased"])
```

```python
df2.head()
```

```python
df[['Gender','Purchased']]=df2[['Gender_encoder','Purchased_encoder']].copy()
df.head()
```
